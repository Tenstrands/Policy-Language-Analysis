```python
import pandas as pd
import pdfkit
import os
```

```python
file_path = "District Level Master Spreadsheet-UCB Policy Analysis - AR 3514.2.xlsx"
df = pd.read_excel(file_path)
```

```python
df
```

```python
# Columns used for naming PDFs
district_col = "District Name"  
website_col = "AR 3514.2 Link to Policy"

output_dir = "generated_pdfs"
os.makedirs(output_dir, exist_ok=True)
```

```python
wkhtmltopdf_path = "/usr/local/bin/wkhtmltopdf"  # macOS default
config = pdfkit.configuration(wkhtmltopdf=wkhtmltopdf_path)
```

```python
def generate_pdf(url, output_path):
    try:
        pdfkit.from_url(url, output_path, configuration=config)
        return output_path
    except Exception as e:
        print(f"Error generating PDF for {url}: {e}")
        return None
```

```python
for index, row in df.iterrows():
    district_name = str(row.get(district_col, "unknown")).replace(" ", "_")
    website_link = row.get(website_col)

    if pd.notna(website_link):  # Ensure the URL is not NaN
        # Format the file name
        file_name = f"{district_name}_AR_3514.2.pdf"
        file_path = os.path.join(output_dir, file_name)

        # Generate the PDF
        pdf_path = generate_pdf(website_link, file_path)
        if pdf_path:
            df.at[index, "AR3514.2 Path to PDF"] = pdf_path  # Save the PDF path in the DataFrame
            
# Save the updated DataFrame to a new Excel file
output_file = "updated_excel_with_pdfs.xlsx"
df.to_excel(output_file, index=False)

print(f"PDFs saved in {output_dir} and Excel updated as {output_file}")
```
