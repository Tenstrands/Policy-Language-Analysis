### Step 4: Clean up Text

```python
import re

def fix_word_boundaries(text):
    # Remove hyphenated line breaks (e.g., "edu-\n" becomes "education")
    text = re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)
    
    # Remove regular line breaks (e.g., "policy\nis" becomes "policy is")
    text = re.sub(r'\n', ' ', text)
    
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
def remove_ed_code_references(text):
    # Remove patterns like "ED 12345" or "Ed. Code § 12345"
    text = re.sub(r'(ED|Ed\.? Code|Education Code)\s?\d+', '', text, flags=re.IGNORECASE)
    return text
```

```python
import re

def clean_text(text):
    # Remove extra whitespace and newlines
    text = re.sub(r'\s+', ' ', text).strip()
    
    
    # Remove page numbers (common format: "Page 1 of 5")
    text = re.sub(r'Page \d+ of \d+', '', text, flags=re.IGNORECASE)
    
    # Remove common header/footer patterns (adjust based on your data)
    text = re.sub(r'(Confidential|Draft|Version \d+)', '', text, flags=re.IGNORECASE)
    
    text = fix_word_boundaries(text)  # Fix broken words and line breaks
    text = remove_ed_code_references(text)
    
    return text
```

```python
policy_pdfs['AR3514.2: PDF Text'] = policy_pdfs['AR3514.2: PDF Text'].fillna('').astype(str)
policy_pdfs['Cleaned Text'] = policy_pdfs['AR3514.2: PDF Text'].apply(clean_text)
```
